```python
from tkinter import *
from tkinter import messagebox
from PIL import Image, ImageTk
import requests
from io import BytesIO

# Function to show a message box with job details
def show_job_detail(job_title):
    messagebox.showinfo("Job Detail", f"Details about {job_title}")

# Function to return to the community main page
def back_to_community(community_window):
    for widget in community_window.winfo_children():
        widget.destroy()
    show_community()

# Function to show community details in the same window
def show_community_detail(community_window, community_title):
    for widget in community_window.winfo_children():
        widget.destroy()
    
    back_button = Button(community_window, text="Back", command=lambda: back_to_community(community_window))
    back_button.pack(anchor='w', padx=10, pady=10)
    
    if community_title == "About":
        info = (
            "About Apna:  "
            "APNA is India's largest professional networking platform, helping millions of "
            "users connect with potential employers, find job opportunities, and advance their careers. "
            "Our mission is to empower the workforce by providing access to meaningful employment opportunities "
            "and fostering professional growth. At APNA, we believe in the power of community and collaboration, "
            "bringing together professionals from diverse backgrounds to share knowledge, insights, and experiences. "
            "Our platform offers a wide range of resources, including job listings, professional development courses, "
            "and networking events, all designed to support our users in their career journeys. Whether you're a recent "
            "graduate looking to start your career, an experienced professional seeking new opportunities, or an employer "
            "searching for top talent, APNA is here to help. Join us today and become part of a vibrant and supportive "
            "professional community dedicated to helping you achieve your career goals."
        )
    elif community_title == "Help":
        info = (
            "Help and Support: "
            "At APNA, we are committed to providing exceptional support to our users. Our Help Center is designed to "
            "assist you with any questions or issues you may encounter while using our platform. From account setup and "
            "profile optimization to job search tips and application guidance, our comprehensive resources cover a wide "
            "range of topics to ensure you have the best experience possible. Our support team is dedicated to resolving "
            "your concerns promptly and effectively. If you need any assistance, our support team is here to help you. "
            "Contact us at help@apna.co or call our toll-free number 1800-123-4567. We are available 24/7 to answer your queries "
            "and provide support. You can also visit our Help Center on the APNA website for more information and resources. "
            "In addition to direct support, our community forums are a great place to connect with other users, share tips, "
            "and find answers to common questions. We also offer live chat support during business hours for immediate assistance. "
            "Your success is our priority, and we are here to ensure that you have all the tools and support you need to achieve your career goals with APNA."
        )
    elif community_title == "Services":
        info = (
            "Our Services: "
            "\n1. Job Listings: Access to a wide range of job opportunities across various industries.\n"
            "\n2. Professional Development: Courses and resources to help you advance your career.\n"
            "\n3. Networking: Connect with professionals and employers in your field.\n"
            "\n4. Career Guidance: Expert advice and support to help you achieve your career goals.\n"
            "\nWhy Choose APNA: "
            "\n1. Largest Networking Platform: Connect with millions of users and potential employers.\n"
            "\n2. Comprehensive Support: Access to a wide range of resources and support services.\n"
            "\n3. Community Focus: A supportive community dedicated to helping you succeed.\n"
            "\n4. Proven Success: Thousands of success stories from our users."
        )
    elif community_title == "Team":
        team_info = [
            ("Anushka", "CEO", "anushka.png"),
            ("Sai Venkat", "CTO", "saivenkat.png"),
            ("Sai Ganesh", "COO", "saiganesh.png"),
            ("Venu", "CFO", "venu.png"),
            ("Sahojit", "CMO", "sahojit.png")
        ]
        
        team_frame = Frame(community_window, bg='snow')
        team_frame.pack(expand=True, fill=BOTH)

        for idx, member in enumerate(team_info):
            name, designation, image_file = member
            try:
                member_image = Image.open(image_file)
                member_image = member_image.resize((150, 150), Image.LANCZOS)
                member_photo = ImageTk.PhotoImage(member_image)
            except Exception as e:
                print(f"Error loading image for {name}: {e}")
                member_photo = None

            member_frame = Frame(team_frame, bg='snow', padx=10, pady=10)
            member_frame.grid(row=idx//3, column=idx%3)

            if member_photo:
                member_label = Label(member_frame, image=member_photo, bg='snow')
                member_label.image = member_photo
                member_label.pack()

            name_label = Label(member_frame, text=name, bg='snow', font=("Helvetica", 14, "bold"))
            name_label.pack()

            designation_label = Label(member_frame, text=designation, bg='snow', font=("Helvetica", 12, "italic"))
            designation_label.pack()
        
        return

    detail_label = Label(
        community_window, 
        text=info, 
        bg='snow', 
        wraplength=400, 
        justify=CENTER, 
        font=("Helvetica", 16),
        padx=20,
        pady=20
    )
    detail_label.pack(expand=True)

# Function to open Jobs window
def show_jobs():
    job_window = Toplevel(win)
    job_window.title("Jobs Information")
    job_window.config(bg='snow')

    job_info_left = [
        'Work From Home Jobs',
        'Part Time Jobs',
        'Freshers Jobs',
        'Jobs for Women',
        'Full Time Jobs',
        'Night Shift Jobs',
        'International Jobs'
    ]

    job_info_right = [
        'Jobs By City',
        'Jobs By Department',
        'Jobs By Company',
        'Jobs By Qualification',
        'Others'
    ]

    left_frame = Frame(job_window, bg='snow')
    left_frame.pack(side=LEFT, fill=BOTH, expand=True, padx=10, pady=10)

    right_frame = Frame(job_window, bg='snow')
    right_frame.pack(side=RIGHT, fill=BOTH, expand=True, padx=10, pady=10)

    for job in job_info_left:
        job_label = Label(left_frame, text=job, bg='snow', cursor="hand2")
        job_label.pack(anchor='w', pady=2)
        job_label.bind("<Button-1>", lambda e, j=job: show_job_detail(j))

    for job in job_info_right:
        job_label = Label(right_frame, text=job, bg='snow', cursor="hand2")
        job_label.pack(anchor='w', pady=2)
        job_label.bind("<Button-1>", lambda e, j=job: show_job_detail(j))

# Function to open Community pop-up
def show_community():
    community_window = Toplevel(win)
    community_window.title("Community Information")
    community_window.config(bg='snow')

    community_info = [
        'About',
        'Help',
        'Services',
        'Team'
    ]

    for community in community_info:
        community_label = Label(community_window, text=community, bg='snow', cursor="hand2", font=("Helvetica", 16, "bold"))
        community_label.pack(anchor='w', padx=10, pady=10)
        community_label.bind("<Button-1>", lambda e, c=community: show_community_detail(community_window, c))
        
# Function to create and show Candidate Login window
def show_candidate_login():
    candidate_login_window = Toplevel(win)
    candidate_login_window.title("Candidate Login")
    candidate_login_window.config(bg='lightblue')
    
    Label(candidate_login_window, text="Candidate Login", font=("Arial", 24, "bold"), bg='lightblue').pack(pady=20)
    
    Label(candidate_login_window, text="Username", bg='lightblue').pack(pady=5)
    Entry(candidate_login_window).pack(pady=5)
    
    Label(candidate_login_window, text="Password", bg='lightblue').pack(pady=5)
    Entry(candidate_login_window, show='*').pack(pady=5)
    
    Button(candidate_login_window, text="Submit", bg='white', fg='seagreen').pack(pady=20)
    
    Label(candidate_login_window, text="Or Sign In With", bg='lightblue').pack(pady=10)
    
    # Load social media logos
    social_logos = {
        "Google": "google_logo.png",
        "GitHub": "github_logo.png",
        "Twitter": "twitter_logo.png"
    }
    social_buttons_frame = Frame(candidate_login_window, bg='lightblue')
    social_buttons_frame.pack(pady=10)
    
    def social_login(platform):
        messagebox.showinfo("Social Login", f"Logging in with {platform}...")
        # Add backend integration code here for social login

    for name, logo_path in social_logos.items():
        try:
            img = Image.open(logo_path)
            img = img.resize((50, 50), Image.LANCZOS)
            photo = ImageTk.PhotoImage(img)
        except Exception as e:
            print(f"Error loading image for {name}: {e}")
            photo = None
        
        if photo:
            btn = Button(social_buttons_frame, image=photo, bg='white', cursor="hand2", command=lambda n=name: social_login(n))
            btn.photo = photo  # Keep a reference to prevent garbage collection
            btn.pack(side=LEFT, padx=10)

# Function to create and show Employer Login window
def show_employer_login():
    employer_login_window = Toplevel(win)
    employer_login_window.title("Employer Login")
    employer_login_window.config(bg='lightblue')
    
    Label(employer_login_window, text="Employer Login", font=("Arial", 24, "bold"), bg='lightblue').pack(pady=20)
    
    Label(employer_login_window, text="Username", bg='lightblue').pack(pady=5)
    Entry(employer_login_window).pack(pady=5)
    
    Label(employer_login_window, text="Password", bg='lightblue').pack(pady=5)
    Entry(employer_login_window, show='*').pack(pady=5)
    
    Button(employer_login_window, text="Submit", bg='white', fg='seagreen').pack(pady=20)
    
    Label(employer_login_window, text="Or Sign In With", bg='lightblue').pack(pady=10)
    
    # Load social media logos
    social_logos = {
        "Google": "google_logo.png",
        "GitHub": "github_logo.png",
        "Twitter": "twitter_logo.png"
    }
    social_buttons_frame = Frame(employer_login_window, bg='lightblue')
    social_buttons_frame.pack(pady=10)
    
    def social_login(platform):
        messagebox.showinfo("Social Login", f"Logging in with {platform}...")
        # Add backend integration code here for social login

    for name, logo_path in social_logos.items():
        try:
            img = Image.open(logo_path)
            img = img.resize((50, 50), Image.LANCZOS)
            photo = ImageTk.PhotoImage(img)
        except Exception as e:
            print(f"Error loading image for {name}: {e}")
            photo = None
        
        if photo:
            btn = Button(social_buttons_frame, image=photo, bg='white', cursor="hand2", command=lambda n=name: social_login(n))
            btn.photo = photo  # Keep a reference to prevent garbage collection
            btn.pack(side=LEFT, padx=10)

# Main window setup
win = Tk()
win.title('APNA - Project')
win.config(bg='snow')

# Load the small APNA logo
try:
    apna_logo_image = Image.open('apna_logo.png')  # Replace with your local path to the image
    apna_logo_image = apna_logo_image.resize((50, 50), Image.LANCZOS)  # Resize the image to a smaller size
    apna_logo_photo = ImageTk.PhotoImage(apna_logo_image)
    print("APNA logo loaded and resized successfully.")
except Exception as e:
    print(f"Error loading or resizing APNA logo: {e}")
    apna_logo_photo = None

# APNA Logo in the top left
if apna_logo_photo:
    apna_logo_label = Label(win, image=apna_logo_photo, bg='snow')
    apna_logo_label.image = apna_logo_photo  # Keep a reference to prevent garbage collection
    apna_logo_label.place(x=10, y=10)

# Create header labels
header_frame = Frame(win, bg='snow')
header_frame.place(x=70, y=30)

# Jobs header
jobs_label = Label(header_frame, text="Jobs", fg='seagreen', bg='snow', font=("Arial", 12, "bold"), cursor="hand2")
jobs_label.pack(side=LEFT, padx=20)
jobs_label.bind("<Button-1>", lambda e: show_jobs())

# Community header
community_label = Label(header_frame, text="Community", fg='seagreen', bg='snow', font=("Arial", 12, "bold"), cursor="hand2")
community_label.pack(side=LEFT, padx=20)
community_label.bind("<Button-1>", lambda e: show_community())

# Create a frame for the search section
f1 = Frame(win, borderwidth=2, bg="white", relief=SUNKEN)
f1.place(x=10, y=350)

# Create a frame for the login buttons
f2 = Frame(win, borderwidth=2, bg="snow", relief=SUNKEN)
f2.pack(side=RIGHT, anchor='ne')

def info1():
    show_candidate_login()

def info2():
    show_employer_login()

# Candidate Login button
b1 = Button(f2, bg='white', fg='seagreen', text="Candidate Login", command=info1)
b1.pack(side=RIGHT, padx=10, pady=10)

# Employer Login button
b2 = Button(f2, bg='white', fg='seagreen', text="Employer Login", command=info2)
b2.pack(side=RIGHT, padx=10, pady=10)

# Load the image from URL
image_url = "https://apna.co/_next/image?url=https%3A%2F%2Fstorage.googleapis.com%2Fmumbai_apnatime_prod%2Fapna-home%2Fnaveen-hero-homepage.png&w=3840&q=75"  # Replace with your image URL
try:
    response = requests.get(image_url)
    image_data = response.content
    image = Image.open(BytesIO(image_data))
    image = image.resize((650, 650), Image.LANCZOS)  # Resize the image to be smaller
    image_photo = ImageTk.PhotoImage(image)
    print("Image loaded successfully.")
except Exception as e:
    print(f"Error loading image: {e}")
    image_photo = None

# Image to the right of labels
if image_photo:
    label_image = Label(win, image=image_photo, bg="snow")
    label_image.image = image_photo  # Keep a reference to prevent garbage collection
    label_image.place(x=800, y=200)  # Adjust x and y coordinates as needed

# Load the app promotion image
try:
    app_promotion_image = Image.open('app_promotion.png')  # Replace with your local path to the image
    app_promotion_image = app_promotion_image.resize((700, 380), Image.LANCZOS)  # Resize the image to be smaller
    app_promotion_photo = ImageTk.PhotoImage(app_promotion_image)
    print("App promotion image loaded and resized successfully.")
except Exception as e:
    print(f"Error loading or resizing app promotion image: {e}")
    app_promotion_photo = None

# Create a frame for the app promotion image
f3 = Frame(win, bg='snow')
f3.place(x=10, y=420)

# App promotion image
if app_promotion_photo:
    app_promotion_label = Label(f3, image=app_promotion_photo, bg='snow')
    app_promotion_label.image = app_promotion_photo  # Keep a reference to prevent garbage collection
    app_promotion_label.pack(side=LEFT, padx=10)

# Main heading
label_heading = Label(win, text="INDIA'S #1 JOB PLATFORM", fg='seagreen', bg='snow', font=("Arial", 19, "bold"))
label_heading.place(x=10, y=150)

# Paragraph
label_para = Label(win, text="Your job search ends here", fg='black', bg='snow', font=("Arial", 45, "bold"))
label_para.place(x=10, y=200)

# Search label
search = Label(f1, text="Search here for jobs", bg='white')
search.grid(row=0, column=0, padx=5, pady=5)

# Search entry
search_value = StringVar()
search_entry = Entry(f1, textvariable=search_value)
search_entry.grid(row=0, column=1, padx=5, pady=5)

# Search button
def clicked():
    messagebox.showinfo("Search Result", f"Searching for: {search_value.get()}")

Button(f1, text="Search", bg='white', fg='seagreen', command=clicked).grid(row=0, column=2, padx=5, pady=5)

# Discover label
label_disc = Label(win, text="Discover 50 lakh+ career opportunities", fg='black', bg='snow', font=("Arial", 19))
label_disc.place(x=10, y=280)

# Footer animation strip
footer_frame = Frame(win, bg='lightblue', height=30)
footer_frame.pack(side=BOTTOM, fill=X)

footer_text = Label(footer_frame, text="Futurense@Apna@2024", bg='lightblue', fg='black', font=("Arial", 15, "bold"))

# Function to animate the footer text
def animate_footer_text():
    x = footer_text.winfo_x()
    if x > win.winfo_width():
        x = -footer_text.winfo_width()
    footer_text.place(x=x+2, y=0)
    win.after(30, animate_footer_text)

animate_footer_text()

win.mainloop()


```
